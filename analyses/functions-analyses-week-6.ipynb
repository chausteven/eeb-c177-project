## Part 1

```python
def datafunc(filename, cols):
    data = pd.read_csv(filename, usecols = cols)
    stat = input('What statistic do you want to find from the data? (i.e. Max, Min, Avg, Std) ')
    if stat == 'Max' or stat == 'max':
        datamax = data.max(axis=0)
        print(datamax)
        plt.plot(data)
        plt.ylabel('Behavior Score')
        plt.xlabel('Subject #')
        output = 'The {} for the dataset in columns {} is:\n{}.\nHere is the dataset:\n{}'.format(stat,cols,datamax,data)
        extract= open('ExtractedMaxData.txt', 'wt')
        extract.write(output)
        extract.close()
    elif stat == 'Min' or stat == 'min':
        datamin=data.min(axis=0)
        print(datamin)
        plt.plot(data)
        plt.ylabel('Behavior Score')
        plt.xlabel('Subject #')
        output = 'The {} for the dataset in columns {} is:\n{}.\nHere is the dataset:\n{}'.format(stat,cols,datamin,data)
        extract= open('ExtractedMinData.txt', 'wt')
        extract.write(output)
        extract.close()
    elif stat == 'Avg' or stat == 'avg':
        dataavg=data.min(axis=0)
        print(dataavg)
        plt.plot(data)
        plt.ylabel('Behavior Score')
        plt.xlabel('Subject #')
        output = 'The {} for the dataset in columns {} is:\n{}.\nHere is the dataset:\n{}'.format(stat,cols,dataavg,data)
        extract= open('ExtractedAvgData.txt', 'wt')
        extract.write(output)
        extract.close()
    else:
        datastd=data.std(axis=0)
        print(datastd)
        plt.plot(data)
        plt.ylabel('Behavior Score')
        plt.xlabel('Subject #')
        output = 'The {} for the dataset in columns {} is:\n{}.\nHere is the dataset:\n{}'.format(stat,cols,datastd,data)
        extract= open('ExtractedStdData.txt', 'wt')
        extract.write(output)
        extract.close()
        
```

```python
#Created two variables and assigned it to zero so that I can count the number of occurances of "Cat" and "Control"
countcat=0 
countcontrol=0

#Using csv.DictReader to reader the csv file as a dictionary and created a for-loop to iterate each order dictionary. Using the key "TREATMENT", it searches if the value matches "Cat" or "Control", if it does match it will add one to the respective count.
with open('Anti-predator.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile, delimiter=',')
    for row in reader:
        if row['TREATMENT'] == 'Cat':
            countcat+=1
        else:
            countcontrol+=1
            
print('There were {} trials that used cats as predators.'.format(countcat))
print('There were {} trials that used no predators.'.format(countcontrol))
```
